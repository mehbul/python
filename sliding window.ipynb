```python
#Given an array of integers Arr of size N and a number K. Return the maximum sum of a subarray of size K.
#Approach 1:
def maxSum(K,Arr,N):
    start=0
    end=0
    s=0
    m=0
    while end<N:
        s=s+Arr[end]
        if end-start+1<K:
#             s=s+Arr[end]
            end=end+1
        elif end-start+1==K:
#             s=s+Arr[end]
            m=max(m,s)
            s=s-Arr[start]
            start=start+1
            end=end+1
    return m
Arr=[100, 200, 300, 400]
N=len(Arr)
K=4
print(maxSum(K,Arr,N))
```

```python
#Given an array of integers Arr of size N and a number K. Return the maximum sum of a subarray of size K.
#Approach 2:
# O(n) solution in Python3 for finding
# maximum sum of a subarray of size k

# Returns maximum sum in
# a subarray of size k.
def maxSum(arr, n, k):

	# k must be greater
	if (n < k):
	
		print("Invalid")
		return -1
	
	# Compute sum of first
	# window of size k
	res = 0
	for i in range(k):
		res += arr[i]

	# Compute sums of remaining windows by
	# removing first element of previous
	# window and adding last element of
	# current window.
	curr_sum = res
	for i in range(k, n):
	
		curr_sum += arr[i] - arr[i-k]
		res = max(res, curr_sum)

	return res

# Driver code
arr = [1, 4, 2, 10, 2, 3, 1, 0, 20]
k = 4
n = len(arr)
print(maxSum(arr, n, k))

# This code is contributed by Anant Agarwal.
```

```python
#Given an array A[] of size N and a positive integer K, find the first negative integer for each and every window(contiguous subarray) of size K.
#Approach 1:
from collections import deque

def printFirstNegativeInteger(arr,n,k):
    start=0
    end=0
    Di= deque()
    ans= []
    while end<n:
        if arr[end]<0:
            Di.append(arr[end])
        if end-start+1<k:
            end=end+1
        elif end-start+1==k:
            if Di:
                ans.append(Di[0])
            else:
                ans.append(0)
            if arr[start]<0:
                Di.popleft()
            start=start+1
            end=end+1
    for i in ans:
        print(i,end=" ")
arr = [12, -1, -7, 8, -15, 30, 16, 28]
n = len(arr)
k = 3
printFirstNegativeInteger(arr, n, k)
```

```python
#Given an array A[] of size N and a positive integer K, find the first negative integer for each and every window(contiguous subarray) of size K.
#Approach 2:
# Python3 implementation to find the
# first negative integer in every window
# of size k import deque() from collections
from collections import deque

# function to find the first negative
# integer in every window of size k
def printFirstNegativeInteger(arr, n, k):
	
	# A Double Ended Queue, Di that will store
	# indexes of useful array elements for the
	# current window of size k. The useful
	# elements are all negative integers.
	Di = deque()

	# Process first k (or first window)
	# elements of array
	for i in range(k):
		
		# Add current element at the rear of Di
		# if it is a negative integer
		if (arr[i] < 0):
			Di.append(i);
	
	# Process rest of the elements, i.e.,
	# from arr[k] to arr[n-1]
	for i in range(k, n):
		
		# if the window does not have
		# a negative integer
		if (not Di):
			print(0, end = ' ')
		
		# if Di is not empty then the element
		# at the front of the queue is the first
		# negative integer of the previous window
		else:
			print(arr[Di[0]], end = ' ');

		# Remove the elements which are
		# out of this window
		while Di and Di[0] <= (i - k):
			Di.popleft() # Remove from front of queue

		# Add current element at the rear of Di
		# if it is a negative integer
		if (arr[i] < 0):
			Di.append(i);

	# Print the first negative
	# integer of last window
	if not Di:
		print(0)
	else:
		print(arr[Di[0]], end = " ")
	
# Driver Code
if __name__ =="__main__":
	arr = [12, -1, -7, 8, -15, 30, 16, 28]
	n = len(arr)
	k = 3
	printFirstNegativeInteger(arr, n, k);

# This code is contributed by
# chaudhary_19 (Mayank Chaudhary)
```

```python
#Given a word pat and a text txt. Return the count of the occurences of anagrams of the word in the text.
#Approach 1:
def search(txt,pat):
    count_letters={}
    
    for letter in pat:
        if letter not in count_letters:
            count_letters[letter]=0
        count_letters[letter]+=1
        
    total_count=len(count_letters)
    
    k=len(pat)
    
    start,end=0,0
    
    ans=0
    while end<len(txt):
        
        if txt[end] in count_letters:
            count_letters[txt[end]]-=1
            if count_letters[txt[end]]==0:
                total_count-=1
                
        if end-start+1<k:
            end+=1
            
        elif end-start+1==k:
            if total_count==0:
                ans+=1
            if txt[start] in count_letters:
                count_letters[txt[start]]+=1
                if count_letters[txt[start]]==1:
                    total_count+=1
            
            start+=1
            end+=1
    return ans
s="aabaabaa"
p="aaba"
ans=search(s,p)
print(ans)
```

```python
#Given a word pat and a text txt. Return the count of the occurences of anagrams of the word in the text.
#Approach 2:
# Python program to search all
# anagrams of a pattern in a text

MAX=256

# This function returns true
# if contents of arr1[] and arr2[]
# are same, otherwise false.
def compare(arr1, arr2):
	for i in range(MAX):
		if arr1[i] != arr2[i]:
			return False
	return True
	
# This function search for all
# permutations of pat[] in txt[]
def search(pat, txt):

	M = len(pat)
	N = len(txt)

	# countP[]: Store count of
	# all characters of pattern
	# countTW[]: Store count of
	# current window of text
	countP = [0]*MAX

	countTW = [0]*MAX

	for i in range(M):
		(countP[ord(pat[i]) ]) += 1
		(countTW[ord(txt[i]) ]) += 1

	# Traverse through remaining
	# characters of pattern
	for i in range(M,N):

		# Compare counts of current
		# window of text with
		# counts of pattern[]
		if compare(countP, countTW):
			print("Found at Index", (i-M))

		# Add current character to current window
		(countTW[ ord(txt[i]) ]) += 1

		# Remove the first character of previous window
		(countTW[ ord(txt[i-M]) ]) -= 1
	
	# Check for the last window in text
	if compare(countP, countTW):
		print("Found at Index", N-M)
		
# Driver program to test above function	
txt = "BACDGABCDA"
pat = "ABCD"	
search(pat, txt)

# This code is contributed
# by Upendra Singh Bartwal
```

```python
#You are given an array of integers nums,there is a sliding window of size k which is moving from the left of the array to the right. 
#You can only see the k numbers in the window. 
#Each time the sliding window moves right by one position.
from collections import deque

def mxslidingwindow(nums,k):
    ans=[]
    start,end=0,0
    Di= deque()
    while end<len(nums):
        while len(Di)>0 and Di[-1]<nums[end]:
            Di.pop()
        Di.append(nums[end])
        if end-start+1<k:
            end=end+1
        elif end-start+1==k:
            ans.append(Di[0])
            if nums[start]==Di[0]:
                Di.popleft()
            start=start+1
            end=end+1
    for i in ans:
        print(i,end=" ")
nums = [1,3,1,2,0,5]
k = 3
mxslidingwindow(nums,k)
```
